```python
def extract(file_path,max_num):    
    file_path = file_path
    class_0 = []
    class_1 = []
    class_0_full = 0
    class_1_full = 0
    with open(file_path, 'r') as f:
        for line in f:
            if len(class_0) < max_num:
                if line.strip()[-1] == "0":
                    line_p = line.strip()
                    class_0.append(line_p)
            else:
                class_0_full = 1
            if len(class_1) < max_num:
                if line.strip()[-1] == "1":
                    line_p = line.strip()
                    class_1.append(line_p)
            else:
                class_1_full = 1
            if class_0_full + class_1_full == 2:
                break
    return class_0, class_1
```

```python
def get_df(data,attribute,is_real):
    data_p =[]
    for d in data:
        d_s = d.split(",")
        if len(d_s) == len(is_real):
            for i in range(len(d_s)):
                #print(i)
                if is_real[i]:
                    d_s[i] = float(d_s[i])
                else:
                    continue
            data_p.append(d_s)
    df = pd.DataFrame(data=data_p,columns=attribute)
    return df
```

```python
traindata_class0 = get_df(train_class_0,attribute,is_real)
```

```python
traindata_class1 = get_df(train_class_1,attribute,is_real)
```

```python
traindata = pd.concat([traindata_class0,traindata_class1])
```
